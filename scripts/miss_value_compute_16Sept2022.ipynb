# Missing forest

```python

#df1 = df.loc[:, [ 's1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 'y1', 'y2', 'y3', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 't1']]
df1 = df.loc[:, ['ID','ID1','Year','dc3', 'dc4','dc5','dc7', 'dc8', 'dc9', 'y1', 'y2', 'y3','s1',  's3', 's4', 's5', 's6','s6_decile', 's8','t1','t2','t3','t4','t5']]
dt1 = df.copy() #created a copy
df1.head(17).T

#Note: with ID1, t1 is missing for some ID1. so we will do it separately.
```

```python
df1.isnull().sum() * 100 / len(df)
```

```python
for idx in df1["ID1"].unique():
        imputer = MissForest(max_iter=15, n_jobs=-1)
        x_imp = imputer.fit_transform(df1[df1.ID1 == idx])
        df1.loc[df1.ID1 == idx, :] = x_imp
```

```python
df1.isna().sum()
```

```python
df1.to_stata('dforest-k.dta')
```

# MICE imputer

```python
# filling the missing value with mice imputer
from fancyimpute import IterativeImputer
mice_imputer = IterativeImputer(missing_values=np.nan, sample_posterior=False, 
                                max_iter=100, tol=0.001, random_state=0,
                                n_nearest_features=30, initial_strategy='mean')

df3 = df.loc[:, ['ID','ID1','Year','dc3', 'dc4','dc5','dc7', 'dc8', 'dc9',  'y1', 'y2', 'y3','s1',  's3', 's4', 's5', 's6','s6_decile', 's8','t1','t2','t3','t4','t5']]
#dt1 = df1.copy() #created a copy
for idx in df3["ID1"].unique():
   
        mice_imp = mice_imputer.fit_transform(df3[df3.ID1 == idx])
        df3.loc[df3.ID1 == idx, :] = mice_imp


#MICE_imputed = mice_imputer.fit_transform(df1)
#MICE_imputed = pd.DataFrame(mice_imp, columns = df3.columns).round(1)
```

```python
df3.isna().sum()
```

```python
df3.to_stata('dmice-k.dta')
```
